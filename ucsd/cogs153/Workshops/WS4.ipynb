```python
import matplotlib.pyplot as plt
import numpy as np

# Data for each plot
# 1. Visual World Paradigm
ambiguous_times = [1.2, 1.3, 1.4, 1.5, 1.6]  # Example times for ambiguous context
unambiguous_times = [0.8, 0.9, 1.0, 1.1, 1.2]  # Example times for unambiguous context
sentence_indices = range(1, 6)

# 2. Sentence Verification Task
bilingual_times = [0.8, 0.85, 0.9, 0.88, 0.87]  # Faster response times
monolingual_times = [1.0, 1.05, 1.1, 1.08, 1.07]  # Slower response times
task_indices = range(1, 6)

# 3. Self-Paced Reading Task
garden_path_times = [1.4, 1.5, 1.6, 1.7, 1.8]  # Slower reading times
control_times = [1.0, 1.1, 1.2, 1.1, 1.0]  # Faster reading times
reading_indices = range(1, 6)

# 4. N400 Amplitudes
conditions = ['Best', 'Semantically Related', 'Semantically Inappropriate']
n400_amplitudes = [-2, -3, -4]  # Negative values for N400 amplitudes

# Plot 1: Visual World Paradigm
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
plt.plot(sentence_indices, ambiguous_times, label='Ambiguous', marker='o')
plt.plot(sentence_indices, unambiguous_times, label='Unambiguous', marker='o')
plt.title('Visual World Paradigm')
plt.xlabel('Sentence Index')
plt.ylabel('Time to Look at Target (s)')
plt.legend()

# Plot 2: Sentence Verification Task
plt.subplot(2, 2, 2)
plt.plot(task_indices, bilingual_times, label='Bilingual', marker='o')
plt.plot(task_indices, monolingual_times, label='Monolingual', marker='o')
plt.title('Sentence Verification Task')
plt.xlabel('Task Index')
plt.ylabel('Response Time (s)')
plt.legend()

# Plot 3: Self-Paced Reading Task
plt.subplot(2, 2, 3)
plt.plot(reading_indices, garden_path_times, label='Garden-Path Sentences', marker='o')
plt.plot(reading_indices, control_times, label='Control Sentences', marker='o')
plt.title('Self-Paced Reading Task')
plt.xlabel('Reading Index')
plt.ylabel('Reading Time (s)')
plt.legend()

# Plot 4: N400 Amplitudes
plt.subplot(2, 2, 4)
plt.bar(conditions, n400_amplitudes, color=['green', 'orange', 'red'])
plt.title('N400 Amplitudes')
plt.xlabel('Sentence Completion Type')
plt.ylabel('N400 Amplitude (V)')
plt.axhline(0, color='black', linewidth=0.5)

plt.tight_layout()
plt.show()
```
