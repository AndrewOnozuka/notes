```python
import numpy as np
import matplotlib.pyplot as plt

# define x[n]
N = 7
n = np.arange(N)
x = np.array([1 if i<=4 else 0 for i in n])

# --- plot x[n] ---
plt.figure()
plt.stem(n, x)
plt.xlabel("n")
plt.ylabel("x[n]")
plt.title("x[n]")
plt.show()

# --- compute DTFS ---
k = np.arange(N)
a = np.zeros(N, dtype=complex)
for kk in k:
    a[kk] = (1/N)*np.sum( x * np.exp(-1j*2*np.pi*kk*n/N) )

# --- magnitude plot ---
plt.figure()
plt.stem(k, np.abs(a))
plt.xlabel("k")
plt.ylabel("|a_k|")
plt.title("Magnitude of DTFS coefficients")
plt.show()

# --- phase plot ---
plt.figure()
plt.stem(k, np.angle(a))
plt.xlabel("k")
plt.ylabel("phase(a_k) [rad]")
plt.title("Phase of DTFS coefficients")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Signal parameters
N = 12
n = np.arange(N)

# Define x[n]
x = 1 - np.sin(np.pi * n / 4)

# Compute Fourier series coefficients a_k
a_k = np.zeros(N, dtype=complex)
for k in range(N):
    a_k[k] = (1/N) * np.sum(x * np.exp(-1j * 2 * np.pi * k * n / N))

# Plot magnitude
plt.figure(figsize=(10,4))
plt.stem(np.arange(N), np.abs(a_k))  # removed use_line_collection
plt.xlabel('k')
plt.ylabel('|a_k|')
plt.title('Magnitude of Fourier Series Coefficients')
plt.grid(True)

# Plot phase
plt.figure(figsize=(10,4))
plt.stem(np.arange(N), np.angle(a_k))  # removed use_line_collection
plt.xlabel('k')
plt.ylabel('Phase of a_k (radians)')
plt.title('Phase of Fourier Series Coefficients')
plt.grid(True)

plt.show()

plt.show()
```
