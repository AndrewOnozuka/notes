```python
import numpy as np
import matplotlib.pyplot as plt

# Define discrete-time signals
n = np.arange(-5, 25)  # sample range (extra padding)
x = np.zeros_like(n)
h = np.zeros_like(n)

# x[n] = 1 for n = 0..4
x[(n >= 0) & (n <= 4)] = 1

# h[n] = 1 for n = 2..7 and n = 11..16
h[((n >= 2) & (n <= 7)) | ((n >= 11) & (n <= 16))] = 1

# Compute convolution
y = np.convolve(x, h)
n_y = np.arange(2 * n[0], 2 * n[-1] + 1)[:len(y)]

# Plot
plt.figure(figsize=(10, 6))

plt.subplot(3, 1, 1)
plt.stem(n, x, basefmt=" ")
plt.title("x[n]")
plt.xlabel("n")
plt.ylabel("Amplitude")

plt.subplot(3, 1, 2)
plt.stem(n, h, basefmt=" ")
plt.title("h[n]")
plt.xlabel("n")
plt.ylabel("Amplitude")

plt.subplot(3, 1, 3)
plt.stem(n_y, y, basefmt=" ")
plt.title("y[n] = x[n] * h[n]")
plt.xlabel("n")
plt.ylabel("Amplitude")

plt.tight_layout()
plt.show()
```
